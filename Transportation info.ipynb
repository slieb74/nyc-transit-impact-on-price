```python
import pandas as pd 
import numpy as np 
import pandas as pd
import requests
import matplotlib.pyplot as plt
%matplotlib inline
import geopandas
from shapely.geometry import Point
```

```python
citi_13= pd.read_csv('citibike2013.csv')
```

```python
citi_13.head()
```

```python
citi_13=citi_13[['start station id','start station name','start station latitude','start station longitude']]
```

```python
citi_13 =citi_13.drop_duplicates(subset=None, keep='first')
```

```python
citi_13.columns
```

```python
citi_13= citi_13.rename(index=str, columns={'start station id': 'id', 'start station name': 'name','start station latitude':'lat','start station longitude':'long'})
```

```python
citi_13.head()
```

```python
geometry13 = [Point(xy) for xy in zip(citi_13.long, citi_13.lat)]
df13 = citi_13.drop(['long','lat'], axis=1)
crs = {'init': 'epsg:4326'}
gdf13 = geopandas.GeoDataFrame(df13, crs=crs, geometry=geometry13)
```

```python
gdf13.head()
```
